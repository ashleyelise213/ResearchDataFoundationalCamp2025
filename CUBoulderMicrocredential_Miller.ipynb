```python
# Import pandas
import pandas as pd
```

```python
# Import dataset
data = pd.read_csv("Mental_Health_Care_in_the_Last_4_Weeks.csv")
data
```

```python
# Check data type
type(data)
```

```python
# Check variable types
data.dtypes
```

```python
data.groupby('Indicator').count()
```

```python
## Filter data

# Filter out NAs in the Value column
data_filtered = data.dropna(subset=['Value'])

# Filter Indicator to only include indicator of interest
data_filtered = data_filtered[data_filtered['Indicator'].isin(['Needed Counseling or Therapy But Did Not Get It, Last 4 Weeks'])]

# Drop unwanted columns
data_filtered = data_filtered[['Indicator', 'Group', 'State', 'Subgroup','Value', 'LowCI', 'HighCI']]
```

```python
data_filtered
```

```python
# Group by Subgroup and calculate mean value per subgroup. Convert to df
data_filtered_mean = data_filtered.groupby('Subgroup')['Value'].mean().reset_index()
```

```python
data_filtered_mean
```

```python
# Sort by Value and display the top five subgroups with the highest mean value
top5 = data_filtered_mean.sort_values(by = 'Value', ascending = False)[0:5]
```

```python
# Sort by Value and display the top five subgroups with the lowest mean value
bottom5 = data_filtered_mean.sort_values(by = 'Value', ascending = True)[0:5]
```

```python
# Combine datasets to plot together
combined_data = pd.concat([top5, bottom5])
combined_data

type(combined_data)
combined_data.dtypes
```

```python
# Import libraries for plotting
import matplotlib.pyplot as plt
```

```python
# Plot mean values of highest and lowest subgroups
combined_data.plot(
    kind = 'bar', # horizontal bar chart
    x = 'Subgroup',
    y = 'Value',
    color=['#1f77b4'],  # Custom color
    title = 'Subgroups of the five highest and lowest mean value'
)

plt.xlabel('Subgroup')
plt.ylabel('Mean Value (%)')
plt.show()
```

### Conclusion

The five subgroups who exhibited the **highest** mean value and answered "Needed Counseling or Therapy But Did Not Get It, Last 4 Weeks" were:
1. Transgender
2. Bisexual
3. Experienced symptoms of anxiety/depression in past 4 weeks
4. With disability
5. Gay or lesbian

The five subgroups who exhibited the **lowest** mean value and answered "Needed Counseling or Therapy But Did Not Get It, Last 4 Weeks" were:
1. 70-79 years
2. 80 years and above
3. Did not experience symptoms of anxiety/depression in past 4 weeks
4. 60-69 years
5. Non-Hispanic Asian, single race
